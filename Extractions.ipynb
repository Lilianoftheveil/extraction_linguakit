AUTOMATIZAÇÃO
   
   Para a investigação das questões 2, 3 e 4 desenvolvi alguns algoritimos para automatizar a extração dos dados.
   Os algoritimos em questão, servem para extrair entidades nomeadas (PESSOAS, LUGARES, ORGANIZAÇÕES e OUTROS), como também lemmas (NOMES, SUBSTANTIVOS, ADJETIVOS e VERBOS), seja por ordem/quantidade de ocorrência ou não.
   
link do repositório com os scripts de extração automática: https://github.com/Lilianoftheveil/extraction_linguakit.git

EXECUTANDO O SCRIPT

   O script "extraction_entity.py", por exemplo, permite a extração de todas entidades de um texto, ou apenas de algumas entidades específicas (PESSOAS, LUGARES, ORGANIZAÇÕES, OUTROS).
   Para sua execução é necessário antes executar o seguinte comando sobre um ficheiro de texto:
   
$ linguakit tagger pt nome_do_texto.txt > nome_do_texto.tag
- este comando identifica todas as palavras / caracteres do texto, e lhes atribui uma entidades e/ou um lemma

    Após isso executa-se o script no terminal referindo-se o arquivo em que se pretende extrair as entidades, no caso "Amor_de_Perdicao.tag":

O comando "help" apresenta todos os comandos - funções - do script:

O comando "entity_person" - por exemplo - serve para a extração das entidades referebtes a pessoas

A seguir podemos optar pela extração de todas as entidades referentes à PESSOAS por ordem ou não de ocorrência.

Por fim, após fazermos mais uma opção, podemos optar por salvar ou não o output, em formato txt ou csv na máquina.
